```python
import csv
import os
import math

# ===== CONFIGURATION =====
# File paths for the exported CSV files.
undist_file = r"C:\Users\abhilash.durgam\Documents\ascc_parts\selected_objects_export_undist.csv"
dist_file   = r"C:\Users\abhilash.durgam\Documents\ascc_parts\selected_objects_export_dist.csv"

# Output file for the deformation results.
output_file = r"C:\Users\abhilash.durgam\Documents\ascc_parts\deformation_experiment.csv"

# ===== UTILITY FUNCTIONS FOR VECTOR OPERATIONS =====
def vector_sub(v1, v2):
    """
    Subtract two 3D vectors represented as tuples.
    Returns a new tuple representing v1 - v2.
    """
    return (v1[0] - v2[0], v1[1] - v2[1], v1[2] - v2[2])

def vector_length(v):
    """
    Compute the Euclidean length of a 3D vector (tuple).
    """
    return math.sqrt(v[0]**2 + v[1]**2 + v[2]**2)

# ===== FUNCTION TO LOAD CSV DATA =====
def load_csv_data(filepath):
    """
    Loads vertex data from a CSV file.
    Assumes CSV headers: "Object Name", "Vertex Index", "WorldX", "WorldY", "WorldZ".
    Returns a dictionary with keys (object_name, vertex_index) and values as a tuple (x, y, z).
    """
    data = {}
    with open(filepath, 'r', newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            obj_name = row["Object Name"]
            index = int(row["Vertex Index"])
            world_x = float(row["WorldX"])
            world_y = float(row["WorldY"])
            world_z = float(row["WorldZ"])
            co = (world_x, world_y, world_z)
            data[(obj_name, index)] = co
    return data

# ===== LOAD THE CSV DATA =====
print("Loading undistorted data from:", undist_file)
undist_data = load_csv_data(undist_file)
print("Loaded", len(undist_data), "vertices from the undistorted file.")

print("Loading distorted data from:", dist_file)
dist_data = load_csv_data(dist_file)
print("Loaded", len(dist_data), "vertices from the distorted file.")

# ===== COMPUTE DEFORMATION PER VERTEX AND WRITE OUTPUT =====
with open(output_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    # Write header row.
    writer.writerow([
        "Object Name", "Vertex Index",
        "Undist_X", "Undist_Y", "Undist_Z",
        "Dist_X", "Dist_Y", "Dist_Z",
        "Delta_X", "Delta_Y", "Delta_Z",
        "Distance"
    ])
    
    # Dictionary to store per-object deformation distances for statistics.
    object_stats = {}
    
    # Loop over each vertex in the undistorted data.
    for key, undist_co in undist_data.items():
        obj_name, index = key
        if key in dist_data:
            dist_co = dist_data[key]
            # Compute the displacement vector and its magnitude.
            delta = vector_sub(dist_co, undist_co)
            distance = vector_length(delta)
            
            # Write the vertex data and computed deformation to the CSV.
            writer.writerow([
                obj_name, index,
                undist_co[0], undist_co[1], undist_co[2],
                dist_co[0], dist_co[1], dist_co[2],
                delta[0], delta[1], delta[2],
                distance
            ])
            
            # Store the distance for statistics.
            if obj_name not in object_stats:
                object_stats[obj_name] = []
            object_stats[obj_name].append(distance)
        else:
            print("Warning: Vertex ({}, {}) not found in distorted data!".format(obj_name, index))

# ===== PRINT DEFORMATION STATISTICS =====
print("\nDeformation Statistics Per Object:")
for obj_name, distances in object_stats.items():
    avg_deformation = sum(distances) / len(distances)
    max_deformation = max(distances)
    print("Object: {}  -  Average Deformation: {:.4f}  |  Max Deformation: {:.4f}".format(
        obj_name, avg_deformation, max_deformation))

print("\nDeformation experiment data written to:", output_file)
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # This registers the 3D projection, no direct call needed
import numpy as np

# ===== CONFIGURATION =====
# Set the path to your deformation experiment CSV file.
csv_path = r"C:\Users\abhilash.durgam\Documents\ascc_parts\deformation_experiment.csv"

# ===== LOAD THE DATA =====
# The CSV is expected to have headers:
# "Object Name", "Vertex Index", "Undist_X", "Undist_Y", "Undist_Z",
# "Dist_X", "Dist_Y", "Dist_Z", "Delta_X", "Delta_Y", "Delta_Z", "Distance"
df = pd.read_csv(csv_path)
print("Loaded {} rows of deformation data.".format(len(df)))

# ===== 3D VISUALIZATION =====
# Create a new figure for a 3D plot.
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the undistorted (base) vertex positions (in blue)
ax.scatter(df['Undist_X'], df['Undist_Y'], df['Undist_Z'],
           color='blue', s=20, alpha=0.6, label='Undistorted')

# Plot the distorted vertex positions (in red)
ax.scatter(df['Dist_X'], df['Dist_Y'], df['Dist_Z'],
           color='red', s=20, alpha=0.6, label='Distorted')

# Optionally, draw arrows (quiver) for the deformation vectors.
# We use a loop here so we can plot an arrow for each vertex.
for i, row in df.iterrows():
    # Starting point of the arrow: the undistorted position
    x, y, z = row['Undist_X'], row['Undist_Y'], row['Undist_Z']
    # The arrow components (delta vector)
    dx, dy, dz = row['Delta_X'], row['Delta_Y'], row['Delta_Z']
    # Plot an arrow. The arrow_length_ratio controls the size of the arrow head.
    ax.quiver(x, y, z, dx, dy, dz, color='green', 
              arrow_length_ratio=0.1, linewidth=0.5)

# Set axis labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("3D Visualization of Deformation Vectors")
ax.legend()

plt.show()

# ===== OPTIONAL: Histogram of Deformation Magnitudes =====
plt.figure(figsize=(8, 5))
plt.hist(df['Distance'], bins=50, color='purple', alpha=0.7)
plt.xlabel("Deformation Magnitude")
plt.ylabel("Frequency")
plt.title("Histogram of Deformation Magnitudes")
plt.show()
```
